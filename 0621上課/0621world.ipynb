```python
import pandas as pd
import numpy as np

#讀入csv資料，只讀特定欄位
df = pd.read_csv("world.csv", index_col='國家',  #用「國家」做為index
                usecols=['國家', '日期','總確診數', '新增確診數','新增死亡數']) #記得還是要把「國家」抓進來
df
```

```python
#取得索引（國家）為台灣的資料
mask = df.index =='台灣'
taiwan = df[mask]
taiwan
```

```python
taiwan[taiwan['新增死亡數'] > 150]
```

```python
#儲存csv檔
taiwan.to_csv('taiwan.csv', encoding='utf-8', index=True)
```
