````python
import pandas as pd

df = pd.read_csv('各鄉鎮市區人口密度.csv') #把csv檔讀成dataframe
df
````

````python
df.info() #資料裡有什麼
````

- RangeIndex：index（索引）有376個（0~375）
- 5個columns（欄位），第0欄有371筆不是null（也就是有5筆是null）
- 類型（Dtype）是字串

````python
#更改columns名稱
df1 = df.rename(columns={'statistic_yyy':'統計年',
                        'site_id':'區域別',
                        'people_total':'年底人口數',
                        'area':'土地面積',
                        'population_density':'人口密度'})
df1
````

````python
#刪除第0列（index 0） 
df2 = df1.drop(index=0) #預設axis=0，axis=0是index，axis=1是column
#刪除有NaN的列
df3 = df2.dropna(axis=0,how='any')
#刪除欄
df4 = df3.drop('統計年',axis=1)
df4
````

````python
df4.info()
````

**用apply才能使用外部function**

````python
#單一欄位是series，要更改單一欄位的資料，要找series的實體方法
#因為人口數中不是每一個都數值，所以用function一個一個處理
def strTOint(x:str)->int:
    try:
        return int(x) #把傳進來的值轉成int數值
    except:
        return 0 #失敗的話，轉為0

df4['年底人口數'] = df4['年底人口數'].apply(strTOint) #apply：使用這個function
df4['人口密度'] = df4['人口密度'].apply(strTOint)

#可以全部一起改的就直接用series的實體方法
df4['土地面積'] = df4['土地面積'].astype('float')

df4.info()
````

````python
#抓資料做新欄位

def split_area(v:str)->str:
    if len(v) >= 6: #如果字數>=6
        return v[:3] #抓前3個字
    else:
        return v #不然直接回傳地區

def split_area1(v:str)->str:
    if len(v) >=6:
        return v[-3:] #抓後3個字
    else:
        return v

df4['縣市'] = df4['區域別'].apply(split_area)
df4['鄉鎮市區'] = df4['區域別'].apply(split_area1)
df4
````

````python
#刪除欄位並重新排列
df5 = df4.reindex(columns=['縣市','鄉鎮市區','年底人口數','土地面積','人口密度'])
df5
````

````python
#搜尋
df5.query('年底人口數>=300000')
````

````python
#最大值
m = df5['人口密度'].max() #最大值放入變數m
df5.query('人口密度 == @m') #在query裡呼叫變數，前面加@
````

````python
#把縣市一樣的合起來，計算各項平均
df5.groupby(by='縣市')[['年底人口數','土地面積','人口密度']].mean() 
````

````python
#用agg同時顯示多個統計量，也可以改顯示名稱
df6 = df5.groupby(by='縣市')[['年底人口數','土地面積']].agg([('平均','mean'),('加總','sum'),'max','min']) 

#加上欄位名稱
df6.columns.names = ["資料", "統計"]

df6
````

````python
#變series
df6.stack(level=['資料','統計'], future_stack=True)
````

````python
#改欄位，改變欄和列 
df6.stack(level=['資料','統計'], future_stack=True).unstack(level=['縣市','統計'])
````

````python
import numpy as np
import pandas as pd 

#美國球賽門票收入
tips_df = pd.read_csv('tips.csv')
tips_df.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips_df['小費比例'] = tips_df['小費'] / tips_df['總票價'] #計算出新的一欄
tips_df
````

````python
grouped = tips_df.groupby(by=['吸煙者','日期']) #合併吸煙者、合併日期
functions = [('數量','count'),('平均','mean'),('最大值','max')] #統計量
tips_df3 = grouped[['小費','總票價']].agg(functions) #計算小費和總票價的統計量
tips_df3
````

````python
tips_df3['小費'] #看全部
tips_df3['小費'][['數量','平均']] #只看數量和平均
````
