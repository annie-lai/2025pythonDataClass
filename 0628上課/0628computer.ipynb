Dataframe：可以當資料庫用

```python
import pandas as pd
df = pd.read_csv('taiwan.csv') #建立出dataframe的實體
type(df)
```

把檔案導入成dataframe後，要先檢查所有欄位和列：
- 欄：直（numpy沒有欄和列的概念，在ndarray是axis=0）
- 列：橫（在ndarray是axis=1）

```python
display(df) #display只有在jupyter裡可以用，直接看df時，預設就是用display顯示
```

```python
df.info()
```

資料的類型：
- str 字串（object）
- int 整數
- float 浮點數
- bool 布林值
- datetime 日值

not-null：沒有空欄

**抓欄位一定是用[]（subscript）**

series：一維資料，類似list和dictionary
要一次抓很多資料：subscript裡放list[[]]

- df.iloc[列,欄]：iloc
- df.loc[列,欄]：loc可以放欄位文字

```python
df[["日期", "新增確診數"]] #欄的選取
df.iloc[:5] #0到5停止，沒有5 #後面不寫就是欄全抓
```

```python
df.loc[:5] #0到5
```

```python
df1 = df.iloc[700:710]
df1
```

```python
df1['新增確診數'] #會跑出series
mask = df1['新增確診數'] >=12 #布林值（True或False）資料
df1[mask]
```

```python
#字串如果是數值或英文，也可以比較大小（中文就不行）
df['日期'] <= '2020-01-20'
```

```python
# 布林值（bool）的資料不能用and，要用&
mask2 = (df['日期'] >= '2020-01-25') & (df['日期'] < '2020-02-01')
df[mask2]
```

```python
#實體方法query（字串） #可以直接用and
#外面用單引號，裡面的字串用雙引號
df.query('日期>"2020-01-20" and 日期<"2020-02-01" and 新增確診數 > 0') 
```

```python
#將字串轉換為datetime
df['日期'] = pd.to_datetime(df['日期']) #直接放df裡的日期型態
df.info()
```

```python
#用日期型式抓特定日期的資料 #型態改成日期就可以用不同的日期型式
df.query('日期 in "2020/01/31"')
```

```python
#將日期做為索引
df_date = df.set_index("日期")
df_date
```

```python
#找2020年一月的資料
df_date.loc['2020/01']
#找2020/1到2020/5的資料
df_date.loc['2020/1':'2020/5']
```
