````python
songs:list[str] = ['songA', 'songB', 'songC', 'songD', 'songE']

import matplotlib.pyplot as plt #import後把matplotlib.pyplot命名為plt

figure = plt.figure() #建立figure實體 #直接用預設()
axes = figure.add_subplot() #建立axes圖表 #預設是(1,1,1)，分成1列row，分成1行col，畫第1個子圖
axes.plot(songs, levels[0],label="hard", color='yellow') #plot(x軸,y軸,樣式) #x和y軸的元素數量要一樣，才能對應
axes.plot(songs, levels[1],label="expert", color='red')
axes.legend()
axes.set_title("hard vs expert",fontdict = {'fontsize': 20}) #字級20
axes.set_xlabel("songs")
axes.set_ylabel('levels')
plt.savefig("songlevel.png") #儲存圖片（檔案名稱要用雙引號） #放在show和close前，不然會被關掉
plt.show()
plt.close()
````

````python
#陣列運算
import math
import numpy as np
from numpy import ndarray 
import matplotlib.pyplot as plt

t:ndarray = np.arange(0, 2, 0.1)
y1:ndarray = np.sin(t * math.pi)
y2:ndarray = np.cos(t * math.pi)
y3:ndarray = t 
figure, axes = plt.subplots() #用triple的方式把figure和axes放入subplots
axes.plot(t, y1, marker='^', linestyle='-.', label='sin')
axes.plot(t, y2, marker='o', linestyle='-', label='cos')
axes.plot(t, y3, marker='*', linestyle='--', label='Θ')
axes.legend(bbox_to_anchor=(1.18, 1)) #加上圖例 #圖例座標
axes.set_title('三角函數')
axes.set_xlabel('角度π')
axes.set_ylabel('值')
plt.show() 
````

````python
import matplotlib.pyplot as plt
#圓餅圖
labels = ["star", "flower", "gress", "moon", "love"]
values = [22, 31, 26, 29, 21]
colors = ["#0B346E","#E87A90", "#90B44B","#6D2E5B", "#FF9409"]
explode = (0, 0.1, 0, 0, 0.05) #不同value要不要有間距

_, _, autotexts = plt.pie(values,labels=labels,colors=colors,explode=explode,startangle=-65,autopct='%1.1f%%') 
#_, _,：前面兩個要放進plt.pie的東西省略
#shadow=要不要有陰影（True）
#startangle=角度
#autopct=百分比的值，1.1：百分比到小數點第一位；多一個%：加上%符號
plt.setp(autotexts, size=14, color='white') #百分比標示的格式
plt.show()
````

````python
#直方圖：數據分佈狀態
import numpy as np
from numpy import ndarray
import matplotlib.pyplot as plt
from scipy.stats import norm #計算常態分佈 #scipy就是numpy的擴充

normal_ditribution:ndarray = np.random.normal(loc=0,scale=1,size=(50))
plt.hist(normal_ditribution, bins=30, color='#B47157',alpha=0.5) #用hist畫直方圖，bin就是分箱（切幾塊），alpha是圖的透明度
# 計算x軸範圍
xmin, xmax = plt.xlim() #xmin和xmax是normal_ditribution的最小值和最大值，plt.xlim()是圖x軸的範圍，預設是0到1
x = np.linspace(xmin, xmax, 100) #在指定範圍（從最小值到最大值）內，總共生成100個等距的數
# 計算常態分佈在特定點 x 的機率密度函數 (Probability Density Function, PDF) 值
# 機率密度：在特定的點上，數據的「集中程度」或「可能性密度」有多高
p = norm.pdf(x, loc=np.mean(normal_ditribution), scale=np.std(normal_ditribution))
plt.plot(x, p, color='red' ,linewidth=2)
print(x)
plt.show()
````
